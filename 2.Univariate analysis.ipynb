### Univariate analysis
- Whether this variable has a significant impact on our process and results
- badrate=bad/total

```python
l=open(r"zodiac_label.txt",encoding="utf-8")
zodiac_label=pd.read_csv(l)
set(zodiac_label.label)# 只有三种标签 0 逾期5天内 1-预期15天以上 2-中甲
```

```python
ft_label=zodiac_label[zodiac_label["label"]!=2]# 
ft_label.shape
ft_label
```

```python
data=pd.merge(ft_label,ft_zodiac,how="inner",on="order_id")
data.head()
```

```python
data.shape
```

```python
zodiac_list=set(data.zodiac)
chinese_zodiac_list=set(data.chinese_zodiac)
```

#### To calculate badrate
- Constellation
- Chinese zodiac

##### Badrate of each constellation
- Store in a dictionary
- Sorted by badrate, and stored in the new dictionary

```python
zodiac_badrate={}
for x in zodiac_list:
    a = data[data.zodiac == x]
    bad=a[a.label == 1]["label"].count()
    good=a[a.label == 0]["label"].count()
    zodiac_badrate[x]=bad/(bad+good)
```

```python
zodiac_badrate
```

```python
f1 = zip(zodiac_badrate.keys(),zodiac_badrate.values())# 将原始统计的键名与值提出来
f1_list=copy.deepcopy(list(f1))
print(f1_list)
f2_list=sorted(f1_list,key= lambda x:x[1],reverse=True)
print(f2_list)
```

```python
zodiac_badrate1=pd.DataFrame(f2_list)
zodiac_badrate1.columns=pd.Series(["constellation","badrate"])
zodiac_badrate1.head()
```

```python
from pyecharts.charts import Line
from pyecharts import options as opts
```

```python
x = zodiac_badrate1["constellation"]
y = zodiac_badrate1["badrate"]
line1=(
    Line()
    .add_xaxis(list(x))
    .add_yaxis("badrate",list(y))
    .set_global_opts(title_opts=opts.TitleOpts(title='Univariate analysis'))
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
)
```

```python
line1.render('pyecharts-line.html')
```

```python
line1.render_notebook()
```
